```python
import requests
```

```python
html_data = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=')
```

```python
content = html_data.text
```

```python

soup = BeautifulSoup(content,'html')
print(soup)
```

```python
company_name = soup.find('h3',class_='joblist-comp-name').text.strip()
experience = soup.find('ul',class_='top-jd-dtl clearfix')
location = experience.find('span').text
experience = experience.li.text.replace('card_travel','')
job_desc = soup.find('ul',class_='list-job-dtl clearfix')
job_description = job_desc.li.text
key_skill = job_desc.span.text.strip().replace(' ,',',')
job_desc_link = soup.find('h2').a['href']

print(f'Company Name is {company_name}')
print(f'Experience is {experience}')
print(f'Location is {location}')
print(f'Job Desc is {job_description}')
print(f'Key skill is {key_skill}')
print(f'Job Desc Link is {job_desc_link}')
```

```python
job_lst = []
def find_job_lst():
    content = html_data.text
    soup = BeautifulSoup(content,'lxml')
    all_jobs = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
    
    for job in all_jobs:
        company_name = job.find('h3',class_='joblist-comp-name').text.strip()
        experience = job.find('ul',class_='top-jd-dtl clearfix')
        location = experience.find('span').text
        experience = experience.li.text.replace('card_travel','')
        job_desc = job.find('ul',class_='list-job-dtl clearfix')
        job_description = job_desc.li.text
        key_skill = job_desc.span.text.strip().replace(' ,',',')
        job_desc_link = job.find('h2').a['href']
        
        my_job_dict = {
            "Company Name":company_name,
            "Experience":experience,
            "Job location":location,
            "Job Desc":job_description,
            "Key Skill":key_skill,
            "Job Desc Link":job_desc_link
            
        }
        
        job_lst.append(my_job_dict)

find_job_lst()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(job_lst)
```

```python
df
```

```python
df.to_csv('Jobs-List.csv')
```

```python
urls_lst = []
url = 'https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page='
for item in range(1,21):
    new_url = url+str(item)
    urls_lst.append(new_url)
```

```python
for links in urls_lst:
    print(links)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
product_data = []

def scrap_all_product():
    for link in urls_lst:
        html_data = requests.get(link)
        content = html_data.text
        soup = BeautifulSoup(content,'lxml')
        all_div = soup.find_all('div',class_='col-sm-4 col-lg-4 col-md-4')
        for item in all_div:
            product_name = item.find('a',class_='title').text
            product_link = item.find('a',class_='title')['href']
            url = 'https://webscraper.io/'
            product_link = url+product_link
            product_price = item.find('h4',class_='pull-right price').text
            product_description = item.find('p',class_='description').text
            reviews = item.find('div',class_='ratings').p.text
            product_dict = {
            'Product Name':product_name,
            'Product Link':product_link,
            'Product Price':product_price,
            'Product Description':product_description,
            'Reviews':reviews
            }
            
            product_data.append(product_dict)
scrap_all_product()
```

```python
import pandas as pd
df_product = pd.DataFrame(product_data)
```
